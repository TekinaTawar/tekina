```
img1=mpimg.imread('/content/abc.jpg')   
img2 =mpimg.imread('/content/abc1.jpg')
print('Image 1 Size:',img1.size)
print('Image 2 Size:',img2.size)
```

```
plt.imshow(img1)
```

```
result1 = img1.reshape([1, img1.size])
result2 = img2.reshape([1, img2.size])
r1=result1.tolist()
r2=result2.tolist()
print(r1)
print(r2)
r1=r1[0]
r2=r2[0]
```

```
def lcs(X, Y,m,n):
  L = [[None]*(n + 1) for i in range(m + 1)]
  for i in range(m + 1):
    for j in range(n + 1):
      if i == 0 or j == 0 :
        L[i][j] = 0
      elif X[i-1] == Y[j-1]:
        L[i][j] = L[i-1][j-1]+1
      else:
        L[i][j] = max(L[i-1][j], L[i][j-1])
  return L[m][n] 
```

```
l= lcs(r1 , r2, len(r1), len(r2))
print("Length of LCS is ",l)
print("Total percentage of similarity :", l/img1.size*100,"%")
```
